## **Rede Perceptron — Porta XOR**

Baseado no exemplo da Profa Sílvia Moraes

> ⚠️ **Atenção:** A porta XOR **não é linearmente separável**. Um único Perceptron **não consegue aprendê-la**. Este notebook demonstra esse limite e por que precisamos de redes com mais camadas (MLP).

```python
import matplotlib.pyplot as plt
import numpy as np
```

Funções de ativação e propagação do sinal

```python
# Funcao de ativacao
def funcaoLimiar(v):
    if v >= 0: return 1
    else: return 0

# Método de propagação
def propagacao(tupla, pesos):
    v = pesos[0] + pesos[1] * tupla[0] + pesos[2] * tupla[1]
    return v
```

Algoritmo de Treinamento: Regra Delta (com limite de épocas)

```python
def algoritmoTreinamento(dados):
    epoca = 0
    eta = 1   # Learning rate = taxa de aprendizado
    pesos = [0, 0, 0]
    print("Pesos iniciais: ", pesos)

    erroGeral = 1
    while erroGeral != 0 and epoca <= 100:
        erroGeral = 0
        epoca = epoca + 1
        print("\nEpoca: ", epoca)
        for tupla in dados:
            print("Processando: ", tupla)
            v = propagacao(tupla, pesos)
            y = funcaoLimiar(v)
            erro = tupla[2] - y
            print('Erro = ', erro)
            if erro != 0:
                delta = eta * erro
                pesos[0] = pesos[0] + delta
                delta = eta * erro * tupla[0]
                pesos[1] = pesos[1] + delta
                delta = eta * erro * tupla[1]
                pesos[2] = pesos[2] + delta
            print("Pesos atuais: ", pesos)
            erroGeral = erroGeral + abs(erro)

    if epoca > 100:
        print("\n*** ATENÇÃO: O treinamento atingiu o limite de épocas sem convergir! ***")
        print("*** O Perceptron simples NÃO consegue aprender XOR. ***")
    return pesos
```

Dados de entrada — Tabela XOR

```python
# Tabela XOR
x1_vals = [0, 0, 1, 1]
x2_vals = [0, 1, 0, 1]
d_vals  = [0, 1, 1, 0]  # XOR: é 1 quando os dois são DIFERENTES

dados = []
print('x1  x2  d')
for i in range(0, len(x1_vals)):
    print(x1_vals[i], ' ', x2_vals[i], ' ', d_vals[i])
    dados.append([x1_vals[i], x2_vals[i], d_vals[i]])
```

Etapa de Treinamento da rede

```python
w = algoritmoTreinamento(dados)
```

Visualização dos pontos XOR — impossível separar com uma reta

```python
w10 = w[0]
w11 = w[1]
w12 = w[2]

x1_d0 = [x1 for x1, d in zip(x1_vals, d_vals) if d == 0]
x2_d0 = [x2 for x2, d in zip(x2_vals, d_vals) if d == 0]
x1_d1 = [x1 for x1, d in zip(x1_vals, d_vals) if d == 1]
x2_d1 = [x2 for x2, d in zip(x2_vals, d_vals) if d == 1]

x = np.linspace(-0.2, 1.2, 100)

if w12 != 0:
    y = (-w10 - w11 * x) / w12
else:
    y = np.full_like(x, np.nan)

plt.figure(figsize=(6, 6))
plt.plot(x, y, 'k--', label=f'Última reta: {w11}·x1 + {w12}·x2 + {w10} = 0')
plt.scatter(x1_d0, x2_d0, color='red', s=100, label='d = 0')
plt.scatter(x1_d1, x2_d1, color='green', s=100, label='d = 1')
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Porta XOR — Não é possível separar com uma reta!')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
print("\nObserve no gráfico: os pontos verdes e vermelhos estão misturados em diagonal.")
print("Nenhuma reta reta consegue separar os dois grupos!")
```

Etapa de Generalização (com os pesos finais — resultado aproximado)

```python
entrada1 = int(input('Informe o valor de x1: '))
entrada2 = int(input('Informe o valor de x2: '))
tuplaEntrada = [entrada1, entrada2]

v = propagacao(tuplaEntrada, w)
y = funcaoLimiar(v)
print('Saida da rede: ', y)
print('(Atenção: este resultado pode estar errado, pois o Perceptron não convergiu para XOR)')
```
