# T1 – Jogo da Velha com Machine Learning
**PUCRS – Inteligência Artificial | Grupo T32**

Classificador de estados do tabuleiro 3×3 em 4 classes:
- `tem_jogo` — partida em andamento
- `x_venceu` — jogador X ganhou
- `o_venceu` — jogador O ganhou
- `empate` — tabuleiro completo sem vencedor

> **Atenção:** Todas as variáveis criadas contêm o código `_32` no nome, conforme especificação do enunciado.

## Seção 0 — Configuração e Constantes

````python
# Cell 0.1 — Importações
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
import warnings
from copy import deepcopy

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, ParameterGrid
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, classification_report, confusion_matrix
)

warnings.filterwarnings('ignore')
%matplotlib inline
print('Bibliotecas carregadas com sucesso.')
````

````python
# Cell 0.2 — Constantes globais (todas com código _32)
RANDOM_STATE_32 = 42
random.seed(RANDOM_STATE_32)
np.random.seed(RANDOM_STATE_32)

CLASS_NAMES_32 = ['tem_jogo', 'x_venceu', 'o_venceu', 'empate']

CELL_COLS_32 = [
    'c1_32', 'c2_32', 'c3_32',
    'c4_32', 'c5_32', 'c6_32',
    'c7_32', 'c8_32', 'c9_32'
]
COLS_32 = CELL_COLS_32 + ['label_32']

# Mapeamento de posições: índice → nome legível
POS_NAMES_32 = [
    'top-left', 'top-mid', 'top-right',
    'mid-left', 'center',  'mid-right',
    'bot-left', 'bot-mid', 'bot-right'
]

ENCODE_MAP_32 = {'x': 1, 'o': -1, 'b': 0}

# 8 linhas vencedoras do jogo da velha
WIN_LINES_32 = [
    (0, 1, 2), (3, 4, 5), (6, 7, 8),   # linhas
    (0, 3, 6), (1, 4, 7), (2, 5, 8),   # colunas
    (0, 4, 8), (2, 4, 6)               # diagonais
]

SAMPLES_PER_CLASS_32 = 200
DATASET_FINAL_PATH_32 = 'tic_tac_toe_balanced_32.csv'

print('Constantes definidas:')
for name in ['RANDOM_STATE_32', 'CLASS_NAMES_32', 'SAMPLES_PER_CLASS_32']:
    print(f'  {name} =', eval(name))
````

## Seção 2 — Engenharia do Dataset

### Problemas encontrados no dataset UCI e soluções aplicadas

| Problema | Solução |
|---|---|
| Apenas 2 classes (`positive`/`negative`) | Re-classificação determinística via lógica do jogo |
| `negative` mistura `o_venceu` e `empate` | Verificação das 8 linhas vencedoras para separar |
| Ausência total de estados `tem_jogo` | Geração sintética via simulação de partidas |
| Dataset desbalanceado (626 vs 332) | Undersampling para ~200 por classe |

````python
# Cell 2.1 — Funções de lógica do tabuleiro (determinísticas, sem ML)

def check_winner_32(board):
    """Verifica se há vencedor. Retorna 'x', 'o' ou None.
    board: lista de 9 strings ('x', 'o', 'b')
    """
    for a, b, c in WIN_LINES_32:
        if board[a] == board[b] == board[c] and board[a] != 'b':
            return board[a]
    return None

def count_pieces_32(board):
    """Retorna tupla (n_x, n_o)."""
    return board.count('x'), board.count('o')

def is_valid_turn_32(board):
    """Valida a ordem dos turnos: n_x - n_o deve ser 0 ou 1 (X inicia)."""
    nx_32, no_32 = count_pieces_32(board)
    return 0 <= (nx_32 - no_32) <= 1

def classify_board_32(board):
    """Classifica o estado de um tabuleiro nas 4 classes do problema.
    board: lista de 9 strings ('x', 'o', 'b')
    Retorna: 'x_venceu' | 'o_venceu' | 'empate' | 'tem_jogo'
    """
    winner_32 = check_winner_32(board)
    if winner_32 == 'x':
        return 'x_venceu'
    if winner_32 == 'o':
        return 'o_venceu'
    if 'b' not in board:   # tabuleiro completo, sem vencedor
        return 'empate'
    return 'tem_jogo'       # casas vazias, sem vencedor

def row_to_board_32(row):
    """Converte linha do DataFrame (colunas CELL_COLS_32) em lista de 9 valores."""
    return [row[col] for col in CELL_COLS_32]

# Testes rápidos das funções
_test_cases_logic_32 = [
    (['x','x','x','o','o','b','b','b','b'], 'x_venceu'),
    (['o','o','o','x','x','b','b','b','b'], 'o_venceu'),
    (['x','o','x','x','o','o','o','x','x'], 'empate'),
    (['x','b','b','b','o','b','b','b','b'], 'tem_jogo'),
]
for _board, _expected in _test_cases_logic_32:
    _result = classify_board_32(_board)
    _ok = '✓' if _result == _expected else '✗'
    print(f'{_ok} classify_board_32({_board[:3]}...) = {_result} (esperado: {_expected})')
print('\nFunções de lógica validadas!')
````

````python
# Cell 2.2 — Relabeling: separar 'negative' em 'o_venceu' e 'empate'

df_32 = df_raw_32.copy()

# Aplicar lógica determinística para reclassificar todas as linhas
df_32['label_32'] = df_32.apply(lambda r: classify_board_32(row_to_board_32(r)), axis=1)

print('=== Distribuição ANTES do relabeling ===')
print(df_raw_32['label_32'].value_counts())
print()
print('=== Distribuição APÓS relabeling (3 classes do UCI) ===')
print(df_32['label_32'].value_counts())

# Sanity check: o dataset UCI é de estados finais — nenhuma linha deve ser 'tem_jogo'
assert (df_32['label_32'] == 'tem_jogo').sum() == 0, \
    'ERRO: estados tem_jogo encontrados no dataset UCI!'
print()
print('✓ Sanity check: 0 estados tem_jogo no dataset UCI (correto — são estados finais)')

# Verificar consistência: positive → x_venceu
n_positive_32 = (df_raw_32['label_32'] == 'positive').sum()
n_x_venceu_32 = (df_32['label_32'] == 'x_venceu').sum()
assert n_positive_32 == n_x_venceu_32, 'ERRO: contagem de x_venceu diverge de positive!'
print(f'✓ Todos os {n_positive_32} registros positive foram mapeados para x_venceu')

# Mostrar quantos negative viraram o_venceu vs empate
n_negative_32 = (df_raw_32['label_32'] == 'negative').sum()
n_o_venceu_32 = (df_32['label_32'] == 'o_venceu').sum()
n_empate_32 = (df_32['label_32'] == 'empate').sum()
print(f'\nDesmembramento dos {n_negative_32} registros negative:')
print(f'  o_venceu: {n_o_venceu_32}')
print(f'  empate:   {n_empate_32}')
````

````python
# Cell 2.3 — Geração sintética de estados 'tem_jogo'
# Estratégia: Random Game Replay
# Simulamos partidas reais (respeitando a ordem de turnos), amostramos
# o tabuleiro em um ponto antes de qualquer vencedor ser declarado.

random.seed(RANDOM_STATE_32)

seen_boards_32 = set()
synthetic_boards_32 = []
target_synthetic_32 = 250   # gerar 250, depois reduzimos para 200
max_attempts_32 = 10000
attempts_32 = 0

while len(synthetic_boards_32) < target_synthetic_32 and attempts_32 < max_attempts_32:
    attempts_32 += 1
    board_s_32 = ['b'] * 9

    # Número de jogadas: 1 a 7 (pelo menos 1 casa vazia, sem tabuleiro cheio)
    n_moves_32 = random.randint(1, 7)
    valid_32 = True

    for turn_32 in range(n_moves_32):
        player_32 = 'x' if turn_32 % 2 == 0 else 'o'
        empty_32 = [i for i, v in enumerate(board_s_32) if v == 'b']
        if not empty_32:
            valid_32 = False
            break
        pos_32 = random.choice(empty_32)
        board_s_32[pos_32] = player_32

        # Se alguém venceu durante a simulação, descartar este percurso
        if check_winner_32(board_s_32) is not None:
            valid_32 = False
            break

    if not valid_32:
        continue

    # Verificar unicidade
    board_tuple_32 = tuple(board_s_32)
    if board_tuple_32 in seen_boards_32:
        continue

    # Sanity check individual
    assert classify_board_32(board_s_32) == 'tem_jogo', \
        f'ERRO: board sintético classificado como {classify_board_32(board_s_32)}'

    seen_boards_32.add(board_tuple_32)
    synthetic_boards_32.append(board_s_32 + ['tem_jogo'])

print(f'Estados tem_jogo gerados: {len(synthetic_boards_32)}')
print(f'Tentativas necessárias:   {attempts_32}')

df_synthetic_32 = pd.DataFrame(synthetic_boards_32, columns=COLS_32)

# Sanity check em lote
assert (df_synthetic_32.apply(lambda r: classify_board_32(row_to_board_32(r)), axis=1) == 'tem_jogo').all(), \
    'ERRO: Nem todos os boards sintéticos são tem_jogo!'
print('✓ Todos os boards sintéticos verificados como tem_jogo')
df_synthetic_32.head()
````

````python
# Cell 2.4 — Montagem e balanceamento do dataset final

df_full_32 = pd.concat([df_32, df_synthetic_32], ignore_index=True)

print('=== Distribuição antes do balanceamento ===')
print(df_full_32['label_32'].value_counts())
print()

# Undersampling: até 200 amostras por classe
# Justificativa: undersampling preserva a qualidade das amostras e evita
# a criação de exemplos artificiais borderline que poderiam degradar o modelo.
dfs_per_class_32 = []
for class_name_32 in CLASS_NAMES_32:
    df_class_32 = df_full_32[df_full_32['label_32'] == class_name_32]
    n_sample_32 = min(SAMPLES_PER_CLASS_32, len(df_class_32))
    df_sampled_32 = df_class_32.sample(n=n_sample_32, random_state=RANDOM_STATE_32)
    dfs_per_class_32.append(df_sampled_32)
    print(f'  {class_name_32}: {len(df_class_32)} disponíveis → {n_sample_32} amostrados')

df_balanced_32 = pd.concat(dfs_per_class_32, ignore_index=True)
df_balanced_32 = df_balanced_32.sample(frac=1, random_state=RANDOM_STATE_32).reset_index(drop=True)

print(f'\nDataset balanceado: {df_balanced_32.shape}')
print(df_balanced_32['label_32'].value_counts())

fig_bal_32, ax_bal_32 = plt.subplots(figsize=(7, 4))
df_balanced_32['label_32'].value_counts().plot(
    kind='bar', ax=ax_bal_32, color=['#2196F3','#4CAF50','#FF9800','#9C27B0'],
    title='Distribuição Final Balanceada'
)
ax_bal_32.set_xlabel('Classe')
ax_bal_32.set_ylabel('Contagem')
ax_bal_32.tick_params(rotation=15)
plt.tight_layout()
plt.show()
````

````python
# Cell 2.5 — Encoding numérico e salvamento

df_encoded_32 = df_balanced_32.copy()

# Encoding das células: x→1, o→-1, b→0
for col_32 in CELL_COLS_32:
    df_encoded_32[col_32] = df_encoded_32[col_32].map(ENCODE_MAP_32)

# Label encoding determinístico
le_32 = LabelEncoder()
le_32.fit(CLASS_NAMES_32)  # ajustado na ordem definida, LabelEncoder ordena alfabeticamente
y_encoded_32 = le_32.transform(df_encoded_32['label_32'])
df_encoded_32['y_32'] = y_encoded_32

print('Mapeamento de classes (LabelEncoder — ordem alfabética):')
for idx_cls, cls_name in enumerate(le_32.classes_):
    print(f'  {idx_cls} → {cls_name}')

# Salvar dataset final
df_encoded_32.to_csv(DATASET_FINAL_PATH_32, index=False)
print(f'\nDataset salvo em: {DATASET_FINAL_PATH_32}')
print(f'Shape: {df_encoded_32.shape}')
df_encoded_32.head()
````

---
## Seção 3 — Pipeline ML: Splits, Scaling e Métricas

````python
# Cell 3.1 — Divisão em treino / validação / teste
# Estratégia: 70% treino | 15% validação | 15% teste
# Justificativa: split fixo e físico garante que todos os 5 algoritmos
# sejam avaliados nos mesmos conjuntos, tornando a comparação justa.
# A validação é usada para tuning de hiperparâmetros; o teste é preservado
# apenas para avaliação final, sem vazamento de informação.

X_32 = df_encoded_32[CELL_COLS_32].values
y_32 = df_encoded_32['y_32'].values

# Primeiro split: 70% treino, 30% temp
X_treino_32, X_temp_32, y_treino_32, y_temp_32 = train_test_split(
    X_32, y_32,
    test_size=0.30,
    stratify=y_32,
    random_state=RANDOM_STATE_32
)

# Segundo split: 50% de temp = 15% validação, 15% teste
X_val_32, X_teste_32, y_val_32, y_teste_32 = train_test_split(
    X_temp_32, y_temp_32,
    test_size=0.50,
    stratify=y_temp_32,
    random_state=RANDOM_STATE_32
)

# Treino + validação combinados (usado para refit do modelo final)
X_trainval_32 = np.vstack([X_treino_32, X_val_32])
y_trainval_32 = np.concatenate([y_treino_32, y_val_32])

print('=== Divisão do Dataset ===')
print(f'Treino:           {X_treino_32.shape[0]:4d} amostras ({X_treino_32.shape[0]/len(X_32)*100:.0f}%) | distribuição: {np.bincount(y_treino_32)}')
print(f'Validação:        {X_val_32.shape[0]:4d} amostras ({X_val_32.shape[0]/len(X_32)*100:.0f}%) | distribuição: {np.bincount(y_val_32)}')
print(f'Teste:            {X_teste_32.shape[0]:4d} amostras ({X_teste_32.shape[0]/len(X_32)*100:.0f}%) | distribuição: {np.bincount(y_teste_32)}')
print(f'Treino+Val:       {X_trainval_32.shape[0]:4d} amostras')
print(f'Total:            {len(X_32):4d} amostras')
````

````python
# Cell 3.2 — Normalização (StandardScaler)
# O scaler é ajustado APENAS no treino para evitar data leakage.
# Tree-based models são invariantes a escala, mas usamos dados escalados
# para todos os modelos, mantendo o pipeline consistente.

scaler_32 = StandardScaler()
X_treino_sc_32 = scaler_32.fit_transform(X_treino_32)
X_val_sc_32 = scaler_32.transform(X_val_32)
X_teste_sc_32 = scaler_32.transform(X_teste_32)
X_trainval_sc_32 = scaler_32.transform(X_trainval_32)

print('Scaler ajustado no treino e aplicado em val/teste.')
print(f'Média (treino, primeiras 3 features): {X_treino_sc_32.mean(axis=0)[:3].round(4)}')
print(f'Std   (treino, primeiras 3 features): {X_treino_sc_32.std(axis=0)[:3].round(4)}')
````

````python
# Cell 3.3 — Funções de avaliação e busca em grid

def avaliar_modelo_32(nome_32, modelo_32, X_test_32, y_test_32):
    """Avalia um modelo treinado e retorna dict com todas as métricas."""
    pred_32 = modelo_32.predict(X_test_32)
    return {
        'nome': nome_32,
        'acuracia_32':        accuracy_score(y_test_32, pred_32),
        'precisao_macro_32':  precision_score(y_test_32, pred_32, average='macro',  zero_division=0),
        'recall_macro_32':    recall_score(y_test_32, pred_32, average='macro',    zero_division=0),
        'f1_macro_32':        f1_score(y_test_32, pred_32, average='macro',        zero_division=0),
        'precisao_por_classe_32': precision_score(y_test_32, pred_32, average=None, zero_division=0),
        'recall_por_classe_32':   recall_score(y_test_32, pred_32, average=None,   zero_division=0),
        'f1_por_classe_32':       f1_score(y_test_32, pred_32, average=None,       zero_division=0),
        'matriz_confusao_32': confusion_matrix(y_test_32, pred_32),
        'modelo_32': modelo_32
    }

def tune_model_32(make_fn_32, param_grid_32, X_tr_32, y_tr_32, X_vl_32, y_vl_32):
    """Busca em grid com conjunto de validação fixo.
    make_fn_32: lambda **p: Estimator(**p)  — permite passar params extras ao construtor
    Retorna: (best_params_dict, best_f1_val, df_log)
    """
    best_f1_32 = -1.0
    best_params_32 = None
    log_32 = []
    warnings.filterwarnings('ignore')

    for params_32 in ParameterGrid(param_grid_32):
        model_32 = make_fn_32(**params_32)
        model_32.fit(X_tr_32, y_tr_32)
        pred_vl_32 = model_32.predict(X_vl_32)
        f1_vl_32 = f1_score(y_vl_32, pred_vl_32, average='macro', zero_division=0)
        log_32.append({**params_32, 'f1_val_32': round(f1_vl_32, 4)})
        if f1_vl_32 > best_f1_32:
            best_f1_32 = f1_vl_32
            best_params_32 = params_32

    warnings.filterwarnings('default')
    df_log_32 = pd.DataFrame(log_32).sort_values('f1_val_32', ascending=False)
    return best_params_32, best_f1_32, df_log_32

print('Funções avaliar_modelo_32 e tune_model_32 definidas.')
````

---
## Seção 4 — Treinamento dos 5 Classificadores

Fluxo padrão para cada algoritmo:
1. Definir grade de hiperparâmetros
2. Busca em grid no conjunto de **validação** (sem tocar no teste)
3. Mostrar top 10 configurações
4. Refit do melhor modelo no conjunto **treino + validação** combinados
5. Modelo final pronto para avaliação no teste

````python
# Cell 4.1 — k-NN (k-Nearest Neighbors)
# Parâmetros a tunear:
#   n_neighbors: número de vizinhos — valores ímpares evitam empates em 2 classes
#   metric: distância euclidiana vs manhattan
#   weights: votos uniformes vs ponderados pela distância inversa

param_grid_knn_32 = {
    'n_neighbors': [1, 3, 5, 7, 9, 11, 15],
    'metric':      ['euclidean', 'manhattan'],
    'weights':     ['uniform', 'distance']
}

print('Buscando melhores hiperparâmetros para k-NN...')
params_knn_32, f1_val_knn_32, grid_knn_32 = tune_model_32(
    lambda **p: KNeighborsClassifier(**p),
    param_grid_knn_32,
    X_treino_sc_32, y_treino_32,
    X_val_sc_32, y_val_32
)

print(f'\nMelhores parâmetros k-NN: {params_knn_32}')
print(f'F1 macro (validação):     {f1_val_knn_32:.4f}')
print('\nTop 10 configurações:')
display(grid_knn_32.head(10))

# Refit no treino + validação
knn_32 = KNeighborsClassifier(**params_knn_32)
knn_32.fit(X_trainval_sc_32, y_trainval_32)
print('\n✓ knn_32 treinado no conjunto treino+val')
````

````python
# Cell 4.2 — Árvore de Decisão
# Parâmetros a tunear:
#   max_depth: profundidade máxima — None = árvore completa (risco de overfitting)
#   criterion: critério de impureza (gini ou entropia)
#   min_samples_split: mínimo de amostras para split
#   min_samples_leaf: mínimo de amostras por folha (regularização)

param_grid_dt_32 = {
    'max_depth':         [None, 3, 5, 7, 10],
    'criterion':         ['gini', 'entropy'],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf':  [1, 2, 4]
}

print('Buscando melhores hiperparâmetros para Árvore de Decisão...')
params_dt_32, f1_val_dt_32, grid_dt_32 = tune_model_32(
    lambda **p: DecisionTreeClassifier(**p, random_state=RANDOM_STATE_32),
    param_grid_dt_32,
    X_treino_sc_32, y_treino_32,
    X_val_sc_32, y_val_32
)

print(f'\nMelhores parâmetros Árvore de Decisão: {params_dt_32}')
print(f'F1 macro (validação):                  {f1_val_dt_32:.4f}')
print('\nTop 10 configurações:')
display(grid_dt_32.head(10))

dt_32 = DecisionTreeClassifier(**params_dt_32, random_state=RANDOM_STATE_32)
dt_32.fit(X_trainval_sc_32, y_trainval_32)
print('\n✓ dt_32 treinado no conjunto treino+val')
````

````python
# Cell 4.3 — MLP (Multilayer Perceptron)
# Parâmetros a tunear:
#   hidden_layer_sizes: topologia — 1 ou 2 camadas ocultas
#   activation: função de ativação (relu vs tanh)
#   alpha: coeficiente de regularização L2
#   max_iter: iterações máximas (fixo em 1000 para garantir convergência)
# Justificativa topológica: entrada = 9 features, 4 classes de saída.
# Redes com 1-2 camadas ocultas são suficientes para este problema estruturado.
# tanh pode se adaptar melhor ao encoding {-1, 0, 1}.

param_grid_mlp_32 = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50), (64, 32)],
    'activation':         ['relu', 'tanh'],
    'alpha':              [0.0001, 0.001, 0.01],
    'max_iter':           [1000]
}

print('Buscando melhores hiperparâmetros para MLP...')
params_mlp_32, f1_val_mlp_32, grid_mlp_32 = tune_model_32(
    lambda **p: MLPClassifier(**p, random_state=RANDOM_STATE_32, early_stopping=False),
    param_grid_mlp_32,
    X_treino_sc_32, y_treino_32,
    X_val_sc_32, y_val_32
)

print(f'\nMelhores parâmetros MLP: {params_mlp_32}')
print(f'F1 macro (validação):    {f1_val_mlp_32:.4f}')
print(f'Topologia usada:         {params_mlp_32["hidden_layer_sizes"]}')
print('\nTop 10 configurações:')
display(grid_mlp_32.head(10))

mlp_32 = MLPClassifier(**params_mlp_32, random_state=RANDOM_STATE_32)
mlp_32.fit(X_trainval_sc_32, y_trainval_32)
print('\n✓ mlp_32 treinado no conjunto treino+val')
````

````python
# Cell 4.4 — Random Forest (Livre escolha #1)
#
# Como funciona: Random Forest é um ensemble de árvores de decisão treinadas
# com bootstrap (amostras com reposição do treino) e seleção aleatória de
# subconjunto de features em cada split. A predição final é por votação
# majoritária das árvores. Isso reduz variância e overfitting em relação a
# uma única árvore de decisão.
#
# Justificativa da escolha: extensão natural da Árvore de Decisão obrigatória;
# permite análise de feature importances que traz interpretabilidade.

param_grid_rf_32 = {
    'n_estimators': [50, 100, 200],
    'max_depth':    [None, 5, 10],
    'max_features': ['sqrt', 'log2']
}

print('Buscando melhores hiperparâmetros para Random Forest...')
params_rf_32, f1_val_rf_32, grid_rf_32 = tune_model_32(
    lambda **p: RandomForestClassifier(**p, random_state=RANDOM_STATE_32),
    param_grid_rf_32,
    X_treino_sc_32, y_treino_32,
    X_val_sc_32, y_val_32
)

print(f'\nMelhores parâmetros Random Forest: {params_rf_32}')
print(f'F1 macro (validação):              {f1_val_rf_32:.4f}')
print('\nTop 10 configurações:')
display(grid_rf_32.head(10))

rf_32 = RandomForestClassifier(**params_rf_32, random_state=RANDOM_STATE_32)
rf_32.fit(X_trainval_sc_32, y_trainval_32)
print('\n✓ rf_32 treinado no conjunto treino+val')
````

````python
# Cell 4.5 — SVM — Support Vector Machine (Livre escolha #2)
#
# Como funciona: SVM encontra o hiperplano de margem máxima que separa as
# classes. Com kernel RBF (Radial Basis Function), transforma o espaço de
# entrada em alta dimensão, permitindo separações não-lineares. O parâmetro C
# controla o trade-off entre margem e erros de classificação. Para problemas
# multiclasse usa estratégia One-vs-Rest.
#
# Justificativa da escolha: abordagem fundamentalmente diferente dos modelos
# baseados em árvores — maximização de margem vs. divisão por impureza.
# O espaço 9D com encoding {-1,0,1} tem estrutura geométrica que SVM explora.

param_grid_svm_32 = {
    'kernel': ['rbf', 'linear', 'poly'],
    'C':      [0.1, 1, 10, 100],
    'gamma':  ['scale', 'auto']
}

print('Buscando melhores hiperparâmetros para SVM...')
params_svm_32, f1_val_svm_32, grid_svm_32 = tune_model_32(
    lambda **p: SVC(**p, decision_function_shape='ovr'),
    param_grid_svm_32,
    X_treino_sc_32, y_treino_32,
    X_val_sc_32, y_val_32
)

print(f'\nMelhores parâmetros SVM: {params_svm_32}')
print(f'F1 macro (validação):    {f1_val_svm_32:.4f}')
print('\nTop 10 configurações:')
display(grid_svm_32.head(10))

svm_32 = SVC(**params_svm_32, decision_function_shape='ovr')
svm_32.fit(X_trainval_sc_32, y_trainval_32)
print('\n✓ svm_32 treinado no conjunto treino+val')
````

### 5.6 — Escolha e Justificativa do Melhor Modelo

O modelo final é selecionado pelo **maior F1 Macro no conjunto de teste**. O F1 Macro é a métrica de escolha porque:
- Leva em conta precision e recall simultaneamente
- É sensível ao desempenho em todas as classes (média simples, não ponderada)
- É mais adequado que acurácia para datasets balanceados com 4 classes

Também verificamos overfitting comparando F1 no treino vs teste. Um gap > 5% é considerado indicativo de overfitting.

````python
# Cell 5.6 — Seleção automática do melhor modelo

melhor_nome_32 = df_comparacao_32['F1_Macro_32'].idxmax()
melhor_resultado_32 = next(r for r in resultados_32 if r['nome'] == melhor_nome_32)
melhor_modelo_32 = melhor_resultado_32['modelo_32']

print(f'=== Melhor Modelo Selecionado: {melhor_nome_32} ===')
print(f'F1 Macro  (teste):  {melhor_resultado_32["f1_macro_32"]:.4f}')
print(f'Acuracia  (teste):  {melhor_resultado_32["acuracia_32"]:.4f}')
print(f'Precisao  (teste):  {melhor_resultado_32["precisao_macro_32"]:.4f}')
print(f'Recall    (teste):  {melhor_resultado_32["recall_macro_32"]:.4f}')

# Verificação de overfitting
pred_tr_best_32 = melhor_modelo_32.predict(X_treino_sc_32)
f1_treino_melhor_32 = f1_score(y_treino_32, pred_tr_best_32, average='macro')
gap_32 = f1_treino_melhor_32 - melhor_resultado_32['f1_macro_32']

print(f'\nF1 Macro  (treino): {f1_treino_melhor_32:.4f}')
print(f'Gap treino-teste:   {gap_32:.4f}')
if gap_32 > 0.05:
    print('⚠ Possivel overfitting detectado (gap > 5%)')
else:
    print('✓ Sem indícios significativos de overfitting')

# Verificar gap com segundo colocado
if len(df_comparacao_32) > 1:
    segundo_32 = df_comparacao_32.index[1]
    diff_32 = df_comparacao_32.loc[melhor_nome_32, 'F1_Macro_32'] - df_comparacao_32.loc[segundo_32, 'F1_Macro_32']
    print(f'\nMargin sobre 2o lugar ({segundo_32}): {diff_32:.4f}')
    if diff_32 < 0.02:
        print('  Diferença pequena (<2%). Considere preferir modelo mais simples.')
````

````python
# Cell 5.7 — Importância das features (se modelo tree-based)

if hasattr(melhor_modelo_32, 'feature_importances_'):
    importances_32 = melhor_modelo_32.feature_importances_
    indices_32 = np.argsort(importances_32)[::-1]

    fig_fi_32, ax_fi_32 = plt.subplots(figsize=(10, 4))
    ax_fi_32.bar(
        range(9),
        importances_32[indices_32],
        color='steelblue',
        edgecolor='white'
    )
    ax_fi_32.set_xticks(range(9))
    ax_fi_32.set_xticklabels([POS_NAMES_32[i] for i in indices_32], rotation=30, ha='right')
    ax_fi_32.set_title(f'Importancia das Features — {melhor_nome_32}')
    ax_fi_32.set_xlabel('Posicao no Tabuleiro')
    ax_fi_32.set_ylabel('Importancia Relativa')
    for i, v in enumerate(importances_32[indices_32]):
        ax_fi_32.text(i, v + 0.002, f'{v:.3f}', ha='center', fontsize=8)
    plt.tight_layout()
    plt.savefig('feature_importance_32.png', dpi=150)
    plt.show()
    print('Figura salva: feature_importance_32.png')
else:
    print(f'Modelo {melhor_nome_32} nao possui feature_importances_.')
    print('Feature importance nao aplicavel para este tipo de modelo.')

    # Para k-NN ou MLP: mostrar permutation importance como alternativa
    from sklearn.inspection import permutation_importance
    print('\nCalculando Permutation Importance como alternativa...')
    perm_32 = permutation_importance(
        melhor_modelo_32, X_teste_sc_32, y_teste_32,
        n_repeats=10, random_state=RANDOM_STATE_32, scoring='f1_macro'
    )
    perm_sorted_idx_32 = perm_32.importances_mean.argsort()[::-1]
    fig_pi_32, ax_pi_32 = plt.subplots(figsize=(10, 4))
    ax_pi_32.bar(
        range(9),
        perm_32.importances_mean[perm_sorted_idx_32],
        color='coral', edgecolor='white'
    )
    ax_pi_32.set_xticks(range(9))
    ax_pi_32.set_xticklabels([POS_NAMES_32[i] for i in perm_sorted_idx_32], rotation=30, ha='right')
    ax_pi_32.set_title(f'Permutation Importance — {melhor_nome_32}')
    ax_pi_32.set_xlabel('Posicao no Tabuleiro')
    ax_pi_32.set_ylabel('Reducao de F1 Macro')
    plt.tight_layout()
    plt.savefig('permutation_importance_32.png', dpi=150)
    plt.show()
````

## Seção 6 — Front End: Jogo Interativo (Humano X CPU Aleatório)

- **Humano** joga como **X** (sempre primeiro)
- **CPU** joga como **O** (movimentos aleatórios)
- Após cada jogada (humano ou CPU), a IA classifica o estado atual
- **Regra especial 1:** se a IA não detectar fim de jogo quando ele ocorreu → jogo é encerrado
- **Regra especial 2:** se a IA detectar fim de jogo erroneamente → log do erro, jogo continua
- Acertos e erros são contabilizados ao longo da partida

````python
# Cell 6.1 — Funções do jogo

def init_board_32():
    """Inicializa tabuleiro vazio."""
    return ['b'] * 9

def display_board_32(board_32):
    """Exibe o tabuleiro formatado. Posições livres mostram o número (1-9)."""
    symbols_32 = {'x': 'X', 'o': 'O'}
    print()
    print(' +---------+')
    for row_32 in range(3):
        row_str_32 = ' | '
        for col_32 in range(3):
            idx_32 = row_32 * 3 + col_32
            if board_32[idx_32] == 'b':
                row_str_32 += str(idx_32 + 1) + ' '
            else:
                row_str_32 += symbols_32[board_32[idx_32]] + ' '
        row_str_32 += '|'
        print(row_str_32)
        if row_32 < 2:
            print(' |---------|')
    print(' +---------+')

def get_cpu_move_32(board_32):
    """Retorna uma posição aleatória entre as células vazias."""
    empty_32 = [i for i, v in enumerate(board_32) if v == 'b']
    return random.choice(empty_32) if empty_32 else None

def board_to_features_32(board_32):
    """Converte board raw para array (1,9) normalizado, pronto para predição."""
    encoded_32 = np.array([ENCODE_MAP_32[v] for v in board_32], dtype=float).reshape(1, -1)
    return scaler_32.transform(encoded_32)

def decode_prediction_32(pred_int_32):
    """Converte inteiro predito de volta para nome da classe."""
    return le_32.inverse_transform([int(pred_int_32)])[0]

def ia_predict_32(board_32):
    """Executa predição da IA para o estado atual do tabuleiro."""
    features_32 = board_to_features_32(board_32)
    pred_int_32 = melhor_modelo_32.predict(features_32)[0]
    return decode_prediction_32(pred_int_32)

print('Funcoes do jogo definidas.')
print(f'IA utiliza o modelo: {melhor_nome_32}')

# Preview do tabuleiro vazio
print('\nVisualização do tabuleiro vazio:')
display_board_32(init_board_32())
````

````python
# Cell 6.2 — Variáveis de tracking de acurácia
total_predicoes_32 = 0
acertos_32 = 0
historico_32 = []   # lista de dicts: um registro por predição da IA

print('Variáveis de tracking inicializadas.')
````

````python
# Cell 6.3 — Loop principal do jogo (executa interativamente)
# Execute esta célula para iniciar uma partida.

def jogar_partida_32():
    global total_predicoes_32, acertos_32, historico_32

    total_predicoes_32 = 0
    acertos_32 = 0
    historico_32 = []

    board_32 = init_board_32()
    turno_32 = 0
    jogo_ativo_32 = True

    print('\n' + '='*45)
    print('  JOGO DA VELHA — Humano (X) vs CPU (O)')
    print('  Digite o número da posição desejada (1-9)')
    print('  Referência:')
    print('  | 1  2  3 |')
    print('  | 4  5  6 |')
    print('  | 7  8  9 |')
    print('='*45)

    def _predizer_e_registrar_32(board_32, jogador_32, turno_32):
        """Executa predição, registra e retorna (true_label, pred_label, correto)."""
        nonlocal total_predicoes_32, acertos_32
        true_label_32 = classify_board_32(board_32)
        pred_label_32 = ia_predict_32(board_32)
        correto_32 = (true_label_32 == pred_label_32)
        total_predicoes_32 += 1
        if correto_32:
            acertos_32 += 1
        historico_32.append({
            'turno_32':      turno_32,
            'jogador_32':    jogador_32,
            'label_real_32': true_label_32,
            'label_pred_32': pred_label_32,
            'correto_32':    correto_32
        })
        acc_atual_32 = acertos_32 / total_predicoes_32
        icon_32 = '✓' if correto_32 else '✗'
        print(f'  [IA] Real: {true_label_32:12s} | Pred: {pred_label_32:12s} | {icon_32} | Acc: {acc_atual_32:.0%}')
        return true_label_32, pred_label_32, correto_32

    while jogo_ativo_32:
        display_board_32(board_32)

        # ── Jogada do Humano ──────────────────────────────────────────────
        while True:
            try:
                entrada_32 = input('\n  Sua jogada (1-9): ').strip()
                pos_32 = int(entrada_32) - 1
                if 0 <= pos_32 <= 8 and board_32[pos_32] == 'b':
                    break
                print('  Posição inválida ou já ocupada. Tente novamente.')
            except ValueError:
                print('  Digite um número inteiro entre 1 e 9.')

        board_32[pos_32] = 'x'
        turno_32 += 1
        print(f'  Humano jogou na posição {pos_32 + 1}')

        true_32, pred_32, ok_32 = _predizer_e_registrar_32(board_32, 'Humano(X)', turno_32)

        if true_32 != 'tem_jogo':
            display_board_32(board_32)
            if not ok_32:
                print(f'\n  ⚠ IA errou: disse "{pred_32}" mas era "{true_32}".')
                print('  Encerrando partida (IA nao detectou fim de jogo).')
            else:
                print(f'\n  ✓ IA detectou corretamente: "{true_32}". Fim de jogo!')
            jogo_ativo_32 = False
            break
        elif pred_32 != 'tem_jogo':
            print(f'  ⚠ IA detectou fim incorretamente ("{pred_32}"). Jogo continua...')

        # ── Jogada do CPU ─────────────────────────────────────────────────
        pos_cpu_32 = get_cpu_move_32(board_32)
        if pos_cpu_32 is None:
            print('  Tabuleiro completo — resultado final acima.')
            break

        board_32[pos_cpu_32] = 'o'
        turno_32 += 1
        print(f'  CPU    jogou na posição {pos_cpu_32 + 1}')

        true_32, pred_32, ok_32 = _predizer_e_registrar_32(board_32, 'CPU(O)', turno_32)

        if true_32 != 'tem_jogo':
            display_board_32(board_32)
            if not ok_32:
                print(f'\n  ⚠ IA errou: disse "{pred_32}" mas era "{true_32}".')
                print('  Encerrando partida (IA nao detectou fim de jogo).')
            else:
                print(f'\n  ✓ IA detectou corretamente: "{true_32}". Fim de jogo!')
            jogo_ativo_32 = False
        elif pred_32 != 'tem_jogo':
            print(f'  ⚠ IA detectou fim incorretamente ("{pred_32}"). Jogo continua...')

# Iniciar a partida
jogar_partida_32()
````

````python
# Cell 6.4 — Relatório pós-partida

if total_predicoes_32 > 0:
    acuracia_interacao_32 = acertos_32 / total_predicoes_32

    print('\n' + '='*50)
    print('  RELATÓRIO DA PARTIDA')
    print('='*50)
    print(f'  Total de predições:          {total_predicoes_32}')
    print(f'  Acertos:                     {acertos_32}')
    print(f'  Erros:                       {total_predicoes_32 - acertos_32}')
    print(f'  Acuracia durante o jogo:     {acuracia_interacao_32:.2%}')
    print(f'  Acuracia no conjunto teste:  {melhor_resultado_32["acuracia_32"]:.2%}')
    print()

    df_historico_32 = pd.DataFrame(historico_32)
    print('Historico detalhado das predições:')
    display(df_historico_32)
else:
    print('Nenhuma partida registrada. Execute a célula anterior primeiro.')
````

## Seção 7 — Verificações de Integridade

````python
# Cell 7.1 — Integridade do dataset
print('=== VERIFICAÇÕES DE INTEGRIDADE DO DATASET ===')
erros_integridade_32 = 0

# 1. Sem valores nulos
nulos_final_32 = df_encoded_32.isnull().sum().sum()
if nulos_final_32 == 0:
    print('✓ Sem valores nulos')
else:
    print(f'✗ Nulos encontrados: {nulos_final_32}')
    erros_integridade_32 += 1

# 2. Balanceamento
contagem_cls_32 = df_encoded_32['label_32'].value_counts()
all_balanced_32 = all(150 <= cnt <= 250 for cnt in contagem_cls_32)
if all_balanced_32:
    print(f'✓ Classes balanceadas: {contagem_cls_32.to_dict()}')
else:
    print(f'✗ Classes fora do intervalo 150-250: {contagem_cls_32.to_dict()}')
    erros_integridade_32 += 1

# 3. Encoding correto
valores_invalidos_32 = []
for col_32 in CELL_COLS_32:
    invalidos = [v for v in df_encoded_32[col_32].unique() if v not in [-1, 0, 1]]
    if invalidos:
        valores_invalidos_32.append(f'{col_32}: {invalidos}')
if not valores_invalidos_32:
    print('✓ Encoding {x:1, o:-1, b:0} correto em todas as colunas')
else:
    print(f'✗ Valores fora de {{-1,0,1}} encontrados: {valores_invalidos_32}')
    erros_integridade_32 += 1

# 4. Duplicatas
n_dupl_32 = df_encoded_32[CELL_COLS_32].duplicated().sum()
print(f'{"✓" if n_dupl_32 == 0 else "!"} Duplicatas no dataset: {n_dupl_32} '
      f'(obs.: tabuleiros iguais com labels diferentes são impossíveis)')

# 5. Consistência labels vs lógica do tabuleiro
erros_label_32 = 0
for _, row_32 in df_balanced_32.iterrows():
    board_check_32 = row_to_board_32(row_32)
    computed_32 = classify_board_32(board_check_32)
    if computed_32 != row_32['label_32']:
        erros_label_32 += 1
if erros_label_32 == 0:
    print('✓ Todas as labels consistentes com lógica do tabuleiro (0 erros)')
else:
    print(f'✗ Labels inconsistentes: {erros_label_32} erros!')
    erros_integridade_32 += 1

print()
if erros_integridade_32 == 0:
    print('✅ Todos os checks de integridade passaram!')
else:
    print(f'⚠ {erros_integridade_32} check(s) falharam. Revise as seções anteriores.')
````

````python
# Cell 7.2 — Sanity checks do modelo
print('=== SANITY CHECKS DO MODELO ===')

_sanity_cases_32 = [
    (['b','b','b','b','b','b','b','b','b'], 'tem_jogo',  'Tabuleiro vazio'),
    (['x','x','x','o','o','b','b','b','b'], 'x_venceu',  'X vence linha superior'),
    (['o','o','o','x','x','b','b','b','b'], 'o_venceu',  'O vence linha superior'),
    (['x','o','x','x','o','o','o','x','x'], 'empate',    'Empate (tabuleiro cheio)'),
    (['x','b','b','b','o','b','b','b','b'], 'tem_jogo',  'Jogo em andamento (2 pecas)'),
]

all_sanity_passed_32 = True
for board_sc_32, expected_32, desc_32 in _sanity_cases_32:
    # Verificação lógica
    true_cls_32 = classify_board_32(board_sc_32)
    logic_ok_32 = (true_cls_32 == expected_32)

    # Verificação do modelo
    feat_sc_32 = board_to_features_32(board_sc_32)
    pred_sc_32 = decode_prediction_32(melhor_modelo_32.predict(feat_sc_32)[0])
    model_ok_32 = (pred_sc_32 == expected_32)

    icon_logic_32 = '✓' if logic_ok_32  else '✗'
    icon_model_32 = '✓' if model_ok_32  else '⚠'
    print(f'{icon_logic_32} Lógica | {icon_model_32} Modelo | [{desc_32}]')
    print(f'  Esperado: {expected_32:12s} | Lógica: {true_cls_32:12s} | Modelo: {pred_sc_32}')

    if not logic_ok_32:
        all_sanity_passed_32 = False

print()
if all_sanity_passed_32:
    print('✅ Todos os sanity checks de lógica passaram!')
else:
    print('⚠ Algum sanity check de lógica falhou — verifique classify_board_32.')
````
